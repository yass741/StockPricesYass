### En utilisant la documentation associée, créer une nouvelle colonne pour chacun des indicateurs suivant.
Les indicateurs sont au format : Nom_de_l_indicateur | Description ou lien vous expliquant la formule | Nom_attendu_de_la_colonne.

 - Relative strenght index | https://en.wikipedia.org/wiki/Relative_strength_index (sur 14 jours) | 'RSI_14'

Vous travaillerez sur à partir de la colonne 'Close' pour l'ensemble de vos formules.


### 1\ Indice de variation

```python
# la methode pct change nous fournit le changement en pourcentage, on met la nouvelle colonne return
df['return'] = df['Close'].pct_change()

df
```

### 2\ Simple Moving Average 7

```python
# afin de calculer le SMA 7 il existe la méthode rolling(window = 7), on calcule la moyenne à chaque 7 jours

sma7 = df['Close'].rolling(window = 7).mean()

plt.plot(sma7, label='SMA 7 jours')
plt.xlabel('Temps')
plt.ylabel('evolution des prix de clôtures')
plt.title('SMA sur 7 jours')
plt.show()
```

### 3\ Simple Moving Average 21

```python
# On refait la même opération mais cette fois-ci avec une valeur de window 21

sma21 = df['Close'].rolling(window = 21).mean()

plt.plot(sma21, label='SMA 21 jours')
plt.xlabel('Temps')
plt.ylabel('evolution des prix de clôtures')
plt.title('SMA sur 21 jours')
plt.show()
```

### 4\ Simple Moving Average Ratio

```python
# on fait le SMA ratio des deux SMA précédents

sma_ratio = sma7/sma21

plt.plot(sma_ratio)
plt.xlabel('Temps')
plt.ylabel('SMA ratio')
plt.title('SMA ratio')
plt.show()
```

### 5\ Relative Strenght Index 14
 - Pour plus de simplicité, utilisez SMA (Simple Moving Average) au lieu de SMMA (Smoothed or modified moving average) dans la formule.

```python
# on commence tout d'abord par définir la différence entre chaque 2 jours consécutifs

diff = df['Close'].diff()

#on définit ensuite le gain et la perte

gain = diff.where(diff > 0, 0)
perte = -diff.where(diff < 0, 0)

#on determine la SMA avec un indice de 14 comme est demandé

moyenne_gain = gain.rolling(14).mean()
moyenne_perte = perte.rolling(14).mean()

# il nous reste maintenant plus que de définir le RSI

rSI = 100 - (100 / (1 + (moyenne_gain / moyenne_perte)))

# maintenat on le plot
plt.plot(rSI)
plt.xlabel("Temps")
plt.ylabel('Relative Strength Index 14')
plt.show()
```

### Affichez la courbe de prix, du SMA_7 et du SMA_21 depuis le 1er septembre 2022

```python
# on selectionne la partie du dataframe a partir de notre date de départ ciblée
new_df = df[df['Date'] >= '2022-09-01']
#on redéfinit les courbes, (je l'ai fait pour plus de temps car le debugger prend plus de temps lorsque j'éxécute tout ce qui est en haut de cette cellule
sma7 = new_df['Close'].rolling(window=7).mean()
sma21 = new_df['Close'].rolling(window=21).mean()
# on plot les différentes courbes
plt.plot(new_df['Date'], new_df['Close'], label='Prix')
plt.plot(new_df['Date'], sma7, label='SMA indice 7')
plt.plot(new_df['Date'], sma21, label='SMA indice 21')
plt.legend()
plt.show()
```
